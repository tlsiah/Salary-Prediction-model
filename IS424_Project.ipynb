# Data Preprocessing
Dealing with outliers and encode categorical variables 

<b> Label Encoding <b>

```python
labelencoder = LabelEncoder()
# Target column:
df['salary_estimate_encoded'] = labelencoder.fit_transform(df['Salary Estimate'])
df.sample(5)
```

<b> One-Hot Encoding <b>

```python
labelbinarizer = LabelBinarizer()
x = labelbinarizer.fit_transform(df["Salary Estimate"])
encoded_df = pd.DataFrame(x, columns=labelbinarizer.classes_)
df = pd.concat([df, encoded_df], axis=1)
df.sample(10)
```
